```python
import numpy as np
import random
```

```python
def sim_samples(num_samples):
    prisoners = list(range(0,7))
    
    correct = []
    
    for i in range(num_samples):
        hats = list(np.random.randint(0,7) for i in range(0,7))
        
        guesses = []
        for i in prisoners:
            sums = (i - (sum(hats[:i]) + sum(hats[i+1:]))) %  7
            guesses.append(sums)
            
        counter = 0
        for i in range(len(guesses)):
            if hats[i] == guesses[i]:
                counter += 1
        
        print(hats)
        print(guesses)
        print()
    
        correct.append(counter)
    
    return correct
```
