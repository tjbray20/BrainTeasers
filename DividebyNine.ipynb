```python
import numpy as np
```

```python
def div_by_9(num):
    counter = 0
    for i in str(num):
        counter += int(i)
    if counter % 9 == 0:
        print(f'{num} is divisible by 9')
    else:
        print(f'{num} is not divisible by 9')
```

```python
for i in range(1000):
    x = np.random.randint(1000000)
    if x % 9 == 0: 
        div_by_9(x)
```
